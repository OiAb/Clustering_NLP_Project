# **Techniques de clustering non supervisé**

Nous commencons par rappeler la définition et l'utilité de l'apprentissage non supervisé.

Nous parlons de l'apprentissage non supervisé quand aucune étiquette n'est donnée à l'algorithme d'apprentissage en essayant de trouver une structure dans son entrée. L'apprentissage non supervisé peut être un objectif en soi (découvrir des modèles cachés dans les données) ou un moyen d'atteindre une fin (apprentissage des fonctionnalités).

L'apprentissage non supervisé peut être classé en deux catégories:


- Apprentissage paramétrique non supervisé :  suppose que les données d'échantillon proviennent d'une population qui suit une distribution de probabilité basée sur un ensemble fixe de paramètres. Théoriquement, dans une famille normale de distributions, tous les membres ont la même forme et sont paramétréspar moyenne et écart type. Cela signifie que si vous connaissez la moyenne et l'écart type, et que la distribution est normale, vous connaissez la probabilité de toute observation future. L'apprentissage paramétrique non supervisé implique la construction de modèles de mélanges gaussiens et l'utilisation de l'algorithme de maximisation des attentes pour prédire la classe de l'échantillon en question. Ce cas est beaucoup plus difficile que l'apprentissage supervisé standard car il n'y a pas d'étiquettes de réponse disponibles et il n'y a donc pas de mesure correcte de l'exactitude disponible pour vérifier le résultat.


- Apprentissage non paramétrique non supervisé : les données sont regroupées en clusters, où chaque cluster (espérons-le) dit quelque chose sur les catégories et les classes présentes dans les données. Cette méthode est couramment utilisée pour modéliser et analyser des données avec des échantillons de petite taille. Contrairement aux modèles paramétriques, les modèles non paramétriques n'exigent pas que le modélisateur fasse des hypothèses sur la distribution de la population et sont donc parfois appelés méthode sans distribution.

Le clustering peut être considéré comme le problème d' apprentissage non supervisé le plus important. Ainsi, comme tout autre problème de ce genre, il s'agit de trouver une structure dans une collection de données non étiquetées. Une définition vague du clustering pourrait être «le processus d'organisation des objets en groupes dont les membres sont similaires d'une certaine manière». Un cluster est donc un ensemble d'objets qui sont «similaires» entre eux et sont «dissemblables» des objets appartenant à d'autres clusters.

clust.png

Les algorithmes de clustering peuvent être classés comme ci-dessous:

- Clustering exclusif : les données sont regroupées de manière exclusive, de sorte que si un certain point de données appartient à un cluster défini, il ne peut pas être inclus dans un autre cluster. K-Means est une méthode appartenant à ce type.

- Clustering par chevauchement : utilise des ensembles flous pour regrouper les données, de sorte que chaque point peut appartenir à deux ou plusieurs clusters avec différents degrés d'appartenance. Dans ce cas, les données seront associées à une valeur d'appartenance appropriée. Nous trouvons comme exemple de méthode de ce genre de clustering Fuzzy C-Means.

- Classification hiérarchique : basé sur l'union entre les deux clusters les plus proches. La condition de début est réalisée en définissant chaque point de données comme un cluster. Après quelques itérations, il atteint les clusters finaux voulus.

- Clustering probabiliste : utilise une approche complètement probabiliste où chaque cluster peut être représenté mathématiquement par une distribution paramétrique, comme une gaussienne. L'ensemble de données est donc modélisé par un mélange de ces distributions. Nous citons la méthode LDA parmi les méthodes appartenant à ce type.

## LDA : Latent Dirichlet Allocation

LDA est un modèle génératif probabiliste permettant d’expliquer des ensembles d’observations, par le moyen de groupes non observés, eux-mêmes définis par des similarités de données, et fait partie d’une catégorie de modèles appelés “topic models” ou modèles de sujet, qui sont un type de modèles de langage statistique utilisés pour découvrir des structures thématiques cachée dans des vastes collections de textes.

LDA suppose que chaque sujet est un mélange sur un ensemble de mots sous-jacent, et chaque document est un mélange de plus d'un ensemble de probabilités de sujet.

image.png

Nous pouvons décrire le processus génératif de LDA comme, étant donné le nombre M de documents, le nombre N de mots et le nombre K préalable de sujets, le modèle s'entraîne à produire:
- La distribution des mots pour chaque sujet
- La répartition des sujets pour chaque document




LDA possède deux hyperparamètres interéssants en plus du paramètre représentant le nombre de sujet :
- Le paramètre alpha est le paramètre de concentration préalable de Dirichlet qui représente la densité des sujets de document - avec un alpha plus élevé, les documents sont supposés être composés de plus de sujets et se traduisent par une distribution plus spécifique des sujets par document.
- Le paramètre bêta est le même paramètre de concentration antérieur qui représente la densité de mots sujet- avec une bêta élevée, les sujets sont supposés être composés de la plupart des mots et se traduisent par une distribution de mots plus spécifique par sujet.

## K-Means

K-Means est une technique de clustering non supervisé très répendu, cette méthode ne contient qu'un seul hyperparamètre K, qui est le nombre de clusters que l'on veut identifier 

Le principe est le suivant :
kmeans.gif



- On place aléatoirement K clusters

- On attribue à chaque point un cluster

- On calcule le barycentre des points par cluster afin de trouver les    coordonnées des nouveaux centroïdes.

- On réattribut à chaque point le cluster dont le centroïde est le plus proche

- On réitère l'opération jusqu'à convergence
  


## Fuzzy C-Means

Dans Fuzzy C-Means, chaque point a une probabilité d'appartenir à chaque cluster, plutôt que d'appartenir complètement à un seul cluster comme c'est le cas dans les k-means traditionnels. Les k-moyens flous essaient spécifiquement de traiter le problème où les points sont quelque peu entre les centres ou autrement ambigus en remplaçant la distance par la probabilité, ce qui pourrait bien sûr être une fonction de la distance, comme avoir une probabilité par rapport à l'inverse de la distance. Les k-moyennes floues utilisent un centre de gravité pondéré basé sur ces probabilités. Les processus d'initialisation, d'itération et de terminaison sont les mêmes que ceux utilisés dans k-means. Les clusters résultants sont mieux analysés comme des distributions probabilistes plutôt que comme une affectation stricte d'étiquettes.

L'algorithme Fuzzy C-Means est le suivant:

- Supposons un nombre fixe de clusters K.

- Initialisation: Initialisez aléatoirement les k-moyennes μk associées aux clusters et calculez la probabilité que chaque point de données Xi soit membre d'un cluster donné K.

- Itération: recalculez le centroïde du cluster en tant que centroïde pondéré étant donné les probabilités d'appartenance à tous les points de données Xi.

```python
stop_words = stopwords.words('french')+['aucun', 'al', 'chez','tant','là','plutôt','autre','dessous','prendre','jusqu','quell','alors','depuis','vouloir','ailleurs','après','désormais','être','où','quelqu','chose','venir','dire','aller','petit','arriver','pouvoir','très','c est','faire', 'pris', 'bref','aller', 'part','sy','jy','d un', 'qu est','nan','ny','n est', 'jai','oui', 'tres','cest','nest','etait','plus','sont','tout','sans','trop','etaient','donc','comme','ca','ete','cela','peut','fait','apres','etre','beaucoup','si','meme','ni','toujours','avoir','jamais','vraiment','car','cet','entre','quon','aussi','fois','netait','elles','dautres','avant','netaient','bien','bon','peu','assez','absolument','surtout','mieux','sympa','rien','bonne','mal']+["dan","san","avoi","mêm","apres","avon","vu","quant","quand","cet","comm","bref","cel","mai","foi","faire"]
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stop_words] for doc in texts]
data_words = list(sent_to_words(dataset))
print(data_words)
# enlever les mots d'arrêt
data_words = remove_stopwords(data_words)
print(data_words)
```

## Évaluation du modèle

La perplexité est l'une des métriques d'évaluation intrinsèques et est largement utilisée pour l'évaluation des modèles de langage. Il saisit à quel point un modèle est surpris de nouvelles données qu'il n'a pas vues auparavant, et est mesuré comme la log-vraisemblance normalisée d'un ensemble de test retenu. Elle doit être petite pour un modèle capable de produire des bon résultats devant des nouvelles données.
 

La cohérence de sujet attribue un score à un seul sujet en mesurant le degré de similitude sémantique entre les mots les mieux notés du sujet. Ces mesures permettent de faire la distinction entre les sujets qui sont des sujets sémantiquement interprétables et les sujets qui sont des artefacts d'inférence statistique. Elle doit donc être élevée pour que le modèle soit cohérent.

Nous calculons le score de cohérence et de perplexité.

```python
# perplexité
print('\nPerplexity : ', lda_model.log_perplexity(corpus))
# coherence
coherence_model_lda = CoherenceModel(model=lda_model, texts=data_words, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence : ', coherence_lda)
```

### Réglage des hyperparamètres

Maintenant que nous avons le score cohérence de base pour le modèle LDA par défaut, effectuons une série de tests de sensibilité pour aider à déterminer les hyperparamètres de modèle suivants:
- Nombre de sujets
- Hyperparamètre de Dirichlet alpha: Densité de sujet de document
- Hyperparamètre de Dirichlet beta: Densité de sujet de mot

Nous effectuerons ces tests en séquence, un paramètre à la fois en gardant les autres constants.

```python
def compute_coherence_values(corpus, dictionary, k, a, b):
    
    lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                           id2word=dictionary,
                                           num_topics=k, 
                                           random_state=100,
                                           chunksize=100,
                                           passes=10,
                                           alpha=a,
                                           eta=b)
    
    coherence_model_lda = CoherenceModel(model=lda_model, texts=data_words, dictionary=id2word, coherence='c_v')
    
    return coherence_model_lda.get_coherence()
```

## Application du modèle LDA

Nous attribuons pour chaque commentaire un sujet dominant.

```python
data=dataset.values.tolist()
def format_topics_sentences(ldamodel=lda_model, corpus=corpus, texts=data):
    sent_topics_df = pd.DataFrame()
    for i, row in enumerate(ldamodel[corpus]):
        row = sorted(row, key=lambda x: (x[1]), reverse=True)
        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0:  # sujet dominant
                wp = ldamodel.show_topic(topic_num)
                topic_keywords = ", ".join([word for word, prop in wp])
                sent_topics_df = sent_topics_df.append(pd.Series([int(topic_num), round(prop_topic*100,1), topic_keywords]), ignore_index=True)
            else:
                break
    sent_topics_df.columns = ['Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords']

    # ajouter le texte original à la fin
    contents = pd.Series(texts)
    sent_topics_df = pd.concat([sent_topics_df, contents], axis=1)
    return(sent_topics_df)

df_topic_sents_keywords = format_topics_sentences(ldamodel=lda_model, corpus=corpus, texts=data)

# format
df_dominant_topic = df_topic_sents_keywords.reset_index()
df_dominant_topic.columns = ['Comment_No', 'Dominant_Topic', 'Topic_Perc_Contrib', 'Keywords', 'Text']
print(df_dominant_topic)
```

Effectuons la méthode elbow entre 1 et 10 clusters, le nombre de cluster optimal est l'abcisse où il y a un coude qui apparaît.
Le coude est entre 5 et 6.

```python
### Methode Elbow
elbow=[]
for k in range(1,8):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(X_train)
    elbow.append([k,kmeans.inertia_])

elbow_=np.array(elbow)
plt.plot(elbow_[:,0],elbow_[:,1])
plt.xlabel('Nombre de clusters')
plt.ylabel('Inertie')
```

Appliquons la silhouette méthode aux alentours de 5 et 6. L'abcisse où la silhouette est maximale, correspondra au nombre de cluster optimal

**Nombre de cluster optimal : 6**

```python
silhouette=[]
for k in range(3,8):
    KMean= KMeans(n_clusters=k)
    KMean.fit(X_train)
    label=KMean.predict(X_train)
    silhouette.append([k,abs(silhouette_score(X_train, label))])

silhouette_=np.array(silhouette)
plt.plot(silhouette_[:,0],silhouette_[:,1])
```

# Méthode Fuzzy C-Means

## Préparation des données pour le modèle Fuzzy C-Means
On utilise CamemBERT pour vectoriser nos phrases.

```python
camembert = torch.hub.load('pytorch/fairseq', 'camembert.v0')
```

```python
def embed(sentence):
  if type(sentence) == list:
    sentence = " ".join(sentence)
  tokens = camembert.encode(sentence)
  # Extract all layer's features (layer 0 is the embedding layer)
  all_layers = camembert.extract_features(tokens, return_all_hiddens=True)
  pooling_layer = all_layers[-2]
  embedded = pooling_layer.mean(1)  # 1 is the dimension you want to average ovber
  # note, using numpy to take the mean is bad if you want to stay on GPU
  return embedded

vectorized_data = []
for word_list in data_words:
  vector = embed(word_list)
  vector = vector.detach().numpy()
  vector = vector.reshape((vector.shape[1],))
  vectorized_data.append(vector)
vectorized_data = np.array(vectorized_data)
vectorized_data
```

## Application du modèle Fuzzy C-Means

On applique le modèle à une phrase. Le seuil permet de déterminer à quels clusters la phrase appartient.
`output[i]` nous dit si la phrase appratient au cluster `i`.

```python
seuil = 1/nb_centers

def fuzzy_cmeans_test(sentence):
  if type(sentence) == list:
    sentence = " ".join(sentence)
  sentence_vectorized = embed(sentence).detach().numpy()
  sentence_output, _, _, _, _, _ = fuzz.cluster.cmeans_predict(sentence_vectorized.T, centers, 2, error=0.005, maxiter=1000)
  return sentence_output > seuil

def fuzzy_cmeans_full_test(id_test):
  test_sentence = data_words[id_test]
  if type(test_sentence) == list:
    test_sentence = " ".join(test_sentence)
  print(test_sentence)
  test_output = fuzzy_cmeans_test(test_sentence)
  return test_output
```
